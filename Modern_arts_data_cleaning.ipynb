```python
import pandas as pd

file = "modern_arts.csv"

data = pd.read_csv(file)
```

```python
data.head(10)
```

```python
data.info()
```

```python
#cleaning nationality column

data["Nationality"] = data["Nationality"].str.replace("(" , "").str.replace(")" , "")
data["Nationality"] = data["Nationality"].str.title()
bol = data["Nationality"] == ''
data.loc[bol , "Nationality"] = "Nationality Unknown"
```

```python
#cleaning Gender column

data["Gender"] = data["Gender"].str.replace("(" , "").str.replace(")" , "")
data["Gender"] = data["Gender"].str.title()
bol2 = data["Gender"] == ''
data.loc[bol2 , "Gender"] = "Gender Unknown"
```

```python
#cleaning dates column

data["BeginDate"] = data["BeginDate"].str.replace("(" , "").str.replace(")" , "").astype(float)

data["EndDate"] = data["EndDate"].str.replace("(" , "").str.replace(")" , "").astype(float)

bad_char = ["c" , "C" , "." , " " , "..." , "s" , "(" , ")"]

def clear_date(date):
    for i in bad_char:
        date = date.replace(i , "")
    return date

```

```python
data["Date"] = data["Date"].apply(clear_date)

```
